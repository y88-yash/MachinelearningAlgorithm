```python
from sklearn import datasets
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn import metrics
```

```python
can=datasets.load_breast_cancer()
```

```python
X_train,X_test,y_train,y_test=train_test_split(can.data,can.target,test_size=0.3,random_state=0)
```

```python
can_gin=DecisionTreeClassifier(criterion='gini',max_depth=5,random_state=0)
```

```python
can_gin.fit(X_train,y_train)
```

```python
pred_gini=can_gin.predict(X_test)
```

```python
print("Model Accuracy with criterion gini",metrics.accuracy_score(y_test,pred_gini))
```

```python
#Overfiting and #Underfiting
print("Trainging set scroe",can_gin.score(X_train,y_train))
print("Test set score",can_gin.score(X_test,y_test))
```

```python
print("Confusion metrix",metrics.confusion_matrix(y_test,pred_gini))
```

```python
print("Classification report ",metrics.classification_report(y_test,pred_gini))
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.figure(figsize=(20,10))
tree.plot_tree(can_gin.fit(X_train,y_train))
```

```python
can_en=DecisionTreeClassifier(criterion='entropy', max_depth=5,random_state=0)
```

```python
can_en.fit(X_train,y_train)
```

```python
pred=can_en.predict(X_test)
```

```python
#over fitting and # under fitting
print("Trainging set score ",can_en.score(X_train,y_train))
print("Test set sCore",can_en.score(X_test,y_test))
```

```python
print("Accuracy of the criterion entropy",metrics.accuracy_score(y_test,pred))
```

```python
print("Confusion metrix",metrics.confusion_matrix(y_test,pred))
```

```python
print("Classification report ",metrics.classification_report(y_test,pred))
```
