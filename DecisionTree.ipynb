```python
from sklearn.datasets import load_wine
```

```python
X,y=load_wine(return_X_y=True)
```

```python
from sklearn import tree
```

```python
clf=tree.DecisionTreeClassifier(min_impurity_decrease=0.2)
```

```python
clf.fit(X,y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(24,14))
tree.plot_tree(clf,filled=True, fontsize=14)
```

```python
clf1=tree.DecisionTreeClassifier(max_depth=3)
clf1.fit(X,y)
```

```python
plt.figure(figsize=(24,14))
tree.plot_tree(clf1,filled=True,fontsize=14)
```

```python
clf3=tree.DecisionTreeClassifier(max_depth=3,min_samples_leaf=3)
```

```python
clf3.fit(X,y)
```

```python
plt.figure(figsize=(24,14))
tree.plot_tree(clf3,filled=True,fontsize=14)
```

```python
clf4=tree.DecisionTreeClassifier(max_leaf_nodes=5)
clf4.fit(X,y)
```

```python
plt.figure(figsize=(20,10))
tree.plot_tree(clf4,filled=True,fontsize=14)
```
